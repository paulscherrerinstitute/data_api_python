```python
import plotly.graph_objs as go
trace = go.Scatter(
    x = data['SINSB01-RIQM-DCP10:FOR-PHASE-AVG'],
    y = data['SINSB01-RKLY-DCP10:FOR-PHASE-AVG'],
    mode='markers')

iplot([trace])
```
